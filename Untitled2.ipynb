# Libraries

```python
import pandas as pd
import numpy as np
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
# for Visualization
from sklearn.decomposition import PCA
import seaborn as sns
```

# Load Dataset

```python
# MovieLens ratings
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=['user_id','item_id','rating','timestamp'])
ratings.sort_values(['user_id','timestamp'], inplace=True)

# Movies info
movies = pd.read_csv('ml-100k/u.item', sep='|', encoding='latin-1', header=None,
                     names=['movie_id','title','release_date','video_release_date','IMDb_URL'] + list(range(19)))
movies['title'] = movies['title'].astype(str)

# Mini dataset for testing
movies = movies.head(100)  # reduce size for notebook test
ratings = ratings[ratings['item_id'].isin(movies['movie_id'])]
```

# Text Embeddings (Transformers)

```python
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
text_model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

def get_text_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = text_model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings.squeeze()

movie_text_embeddings = {}
for idx, row in movies.iterrows():
    movie_text_embeddings[row['movie_id']] = get_text_embedding(row['title'])
```

# Image Embeddings (CNN)

```python
# Dummy images for template
# In real project, load movie posters from files
image_size = (224,224)
transform = transforms.Compose([transforms.Resize(image_size),
                                transforms.ToTensor()])

cnn_model = models.resnet18(pretrained=True)
cnn_model.fc = nn.Identity()  # output feature vector
cnn_model.eval()

def get_image_embedding(image_path):
    image = Image.open(image_path).convert('RGB')
    img_tensor = transform(image).unsqueeze(0)
    with torch.no_grad():
        features = cnn_model(img_tensor)
    return features.squeeze()

# For demo, create random embeddings
movie_image_embeddings = {mid: torch.rand(512) for mid in movies['movie_id']}
```

# Prepare sequences for RNN

```python
seq_len = 5
user_sequences = ratings.groupby('user_id')['item_id'].apply(list).to_dict()

X_seq, X_text, X_image, y_seq = [], [], [], []

for user, seq in user_sequences.items():
    for i in range(len(seq)-seq_len):
        seq_ids = seq[i:i+seq_len]
        X_seq.append(seq_ids)
        y_seq.append(seq[i+seq_len])
        # Text embeddings
        seq_text_emb = torch.stack([movie_text_embeddings[item] for item in seq_ids])
        X_text.append(seq_text_emb)
        # Image embeddings
        seq_img_emb = torch.stack([movie_image_embeddings[item] for item in seq_ids])
        X_image.append(seq_img_emb)

X_text = torch.stack(X_text)
X_image = torch.stack(X_image)
y_seq = torch.tensor(y_seq)
```

# Dataset & DataLoader

```python
class MultiModalDataset(Dataset):
    def __init__(self, X_text, X_image, y):
        self.X_text = X_text
        self.X_image = X_image
        self.y = y
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X_text[idx], self.X_image[idx], self.y[idx]

dataset = MultiModalDataset(X_text, X_image, y_seq)
train_size = int(0.8*len(dataset))
test_size = len(dataset)-train_size
train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size,test_size])
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

# Multi-Modal Model

```python
class MultiModalRecommender(nn.Module):
    def __init__(self, text_dim, image_dim, hidden_size, n_items):
        super().__init__()
        self.rnn_text = nn.GRU(text_dim, hidden_size, batch_first=True)
        self.rnn_image = nn.GRU(image_dim, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size*2, n_items+1)
    
    def forward(self, x_text, x_image):
        out_text, _ = self.rnn_text(x_text)
        out_image, _ = self.rnn_image(x_image)
        out = torch.cat([out_text[:,-1,:], out_image[:,-1,:]], dim=1)
        out = self.fc(out)
        return out

text_dim = X_text.shape[2]
image_dim = X_image.shape[2]
hidden_size = 64
n_items = ratings['item_id'].max()

model = MultiModalRecommender(text_dim, image_dim, hidden_size, n_items)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

# Training Loop

```python
epochs = 2  # for testing
for epoch in range(epochs):
    model.train()
    total_loss = 0
    for x_text, x_img, y in train_loader:
        optimizer.zero_grad()
        outputs = model(x_text, x_img)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(train_loader):.4f}")
```

# Top-N Recommendations

```python
def recommend_top_n(model, x_text, x_img, top_n=5):
    model.eval()
    with torch.no_grad():
        outputs = model(x_text.unsqueeze(0), x_img.unsqueeze(0))
        top_items = outputs[0].topk(top_n).indices.tolist()
    return top_items

example_text = X_text[0]
example_image = X_image[0]
top5 = recommend_top_n(model, example_text, example_image)
print("Top 5 recommended items:", top5)
```

# Visualtization
## 1. Top N predictions

```python
def plot_top_n(top_items, n=5):
    plt.bar(range(1, n+1), np.arange(n,0,-1))
    plt.xticks(range(1, n+1), top_items)
    plt.title(f"Top {n} Recommended Items")
    plt.xlabel("Item ID")
    plt.ylabel("Rank")
    plt.show()

plot_top_n(top5)
```

## 2. Embeddings (PCA / t-SNE)

```python
# Uzmi embedding vektore za sve filmove
all_embeddings = torch.stack(list(movie_text_embeddings.values())).numpy()
movie_ids = list(movie_text_embeddings.keys())

# PCA redukcija na 2D
pca = PCA(n_components=2)
emb_2d = pca.fit_transform(all_embeddings)

# Scatter plot
plt.figure(figsize=(10,7))
plt.scatter(emb_2d[:,0], emb_2d[:,1], alpha=0.7)
for i, mid in enumerate(movie_ids[:20]):  # samo 20 filmova za label
    plt.text(emb_2d[i,0], emb_2d[i,1], str(mid), fontsize=9)
plt.title("2D PCA of Text Embeddings (Transformers)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

## 3. Top-N Heatmap (for more users)

```python
top_n_matrix = []
for i in range(min(10, X_text.shape[0])):  # 10 korisnika
    top_items = recommend_top_n(model, X_text[i], X_image[i])
    top_n_matrix.append(top_items)

top_n_matrix = np.array(top_n_matrix)

plt.figure(figsize=(10,5))
sns.heatmap(top_n_matrix, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Top-N Recommended Items for 10 Users")
plt.xlabel("Top-N rank")
plt.ylabel("User index")
plt.show()
```
